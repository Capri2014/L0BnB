```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def penalty_function(lambda_0, lambda_2, m, beta):
    golden_ratio = np.sqrt(lambda_0/ lambda_2)
    if golden_ratio <= abs(beta):
        return lambda_0 + lambda_2*beta**2
    elif golden_ratio < m:
        return 2*abs(beta)*np.sqrt(lambda_0*lambda_2)
    else:
        return abs(beta)*(lambda_0/m + lambda_2*m)
```

```python
beta = np.linspace(-2, 2, 50000)
p_1 = [penalty_function(9, 1, 2, i) for i in beta]
p_2 = [penalty_function(1, 1, 2, i) for i in beta]
```

```python
plt.plot(beta, p_1)
```

```python
plt.plot(beta, p_2)
```

```python
def beta_update_function(lambda_0, lambda_2, m, xiri):
    golden_ratio = np.sqrt(lambda_0/ lambda_2)
    sign_xiri = np.sign(xiri)
    abs_xiri = abs(xiri)
    if golden_ratio < m:
        if abs_xiri <= 2*np.sqrt(lambda_0*lambda_2):
            return 0
        elif abs_xiri < np.sqrt(lambda_0/lambda_2) + 2*np.sqrt(lambda_0*lambda_2):
            return (abs_xiri - 2*np.sqrt(lambda_0*lambda_2))*sign_xiri
        elif abs_xiri <= m + 2*lambda_2*m:
            return xiri/(1 + 2*lambda_2)
        else:
            return m*sign_xiri
    else:
        if abs_xiri <= lambda_0/m + lambda_2*m:
            return 0
        elif abs_xiri <= m + lambda_0/m + lambda_2*m:
            return (abs_xiri - (lambda_0/m + lambda_2*m))*sign_xiri
        else:
            return  m*sign_xiri
```

```python
xiri = np.linspace(-10, 10, 50000)
updates_1 = [beta_update_function(9, 1, 2, i) for i in xiri]
updates_2 = [beta_update_function(1, 1, 2, i) for i in xiri]
```
